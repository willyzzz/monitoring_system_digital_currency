```python
def get_id(keywords,token_find = ""):
    import pandas as pd
    try:
        id_coin = token_find[token_find.symbol.str.lower() == keywords.lower()].iloc[0,0]
    except:
        id_coin = 0
    return id_coin
```

```python
def get_name(keywords,token_find = ""):
    import pandas as pd
    try:
        name_coin = token_find[token_find.symbol.str.lower() == keywords.lower()].iloc[0,2]
    except:
        name_coin = 0
    return name_coin
```

## 调用Coingecko获取基础数据

```python
def get_cg_data(idx):
    from pycoingecko import CoinGeckoAPI
    cg = CoinGeckoAPI()
    coin_cg = cg.get_coin_by_id(idx)
    cg_rank =  coin_cg["coingecko_rank"]
    cg_score = coin_cg["coingecko_score"]
    market_rank = coin_cg["market_cap_rank"]
    daily_vol = coin_cg['market_data']['total_volume']["usd"]
    try:
        tvl = coin_cg['market_data']['total_value_locked']["usd"]
    except:
        tvl = 0
    try:
        mcap_t_tvl = coin_cg['market_data']['mcap_to_tvl_ratio']
    except:
        mcap_t_tvl = 0
        
    price_change = coin_cg['market_data']['price_change_percentage_30d']
    number_of_exchanges = len(set([i['market']['name'] for i in coin_cg['tickers']]))
    cg_data = [idx, cg_rank, cg_score, market_rank, daily_vol, tvl, mcap_t_tvl, price_change, number_of_exchanges]
    return cg_data 
```

### Coinbase 正数据

```python
cb_data = []
for i in cb_id:
    cb_data.append(get_cg_data(i))
    time.sleep(0.4)
cb_data = pd.DataFrame(cb_data, columns = ["id","cg_rank","cg_score","market_rank","daily_vol","tvl","mcap_t_tvl","price_change","number_of_exchanges"])
```
